```python
import numpy as np
import pandas as pd

# Define the x values
x = np.linspace(0,  np.pi, 1000)

# Calculate the sinus and cosine functions
sin_x = np.sin(x)
cos_x = np.cos(x)
cos_2x = np.cos(2 * x) * 0.5
cos_2xp1 = np.cos(2 * x +1) * 1.5
# Create a pandas DataFrame
data = {
    'x': x,
    'sin(x)': sin_x,
    'cos(x)': cos_x,
    'cos(2x)*0.5': cos_2x,
    'cos(2x +1) * 1.5': cos_2xp1,
}
df = pd.DataFrame(data)
df= df.set_index('x')
# Display the DataFrame
print(df)
```

```python
title = 'Beautiful Line Plot'
substitle = 'Trigonometric Functions in the range 0 to Ï€'
```

```python
import numpy as np

#%% MATPLOT LIB EXAMPLE
from matplotlib import pyplot as plt  # Import matplotlib

plt.style.use('richview.mplstyle')  # Use the richview style

plt.plot(df)  # Plot the data
plt.legend(df.columns, ncol =len(df.columns),loc='upper left', bbox_to_anchor=(0, 1.075))  # Add the legend
plt.suptitle(substitle)  # Add the subtitle
plt.title(title, y =1.15)

plt.show()  # Show the plot
```
