```python
import numpy as np
import pandas as pd

# Define the x values
x = np.linspace(0,  np.pi, 1000)

# Calculate the sinus and cosine functions
sin_x = np.sin(x)
cos_x = np.cos(x)
cos_2x = np.cos(2 * x) * 0.5
cos_2xp1 = np.cos(2 * x +1) * 1.5
# Create a pandas DataFrame
data = {
    'x': x,
    'sin(x)': sin_x,
    'cos(x)': cos_x,
    'cos(2x)*0.5': cos_2x,
    'cos(2x +1) * 1.5': cos_2xp1,
}
df = pd.DataFrame(data)
df= df.set_index('x')
# Display the DataFrame
print(df)

width=800
height=600
```

```python
import numpy as np
title = 'Beautiful Line Plot'
subtitle = 'Trigonometric Functions in the range 0 to Ï€'
#%% MATPLOT LIB EXAMPLE
from matplotlib import pyplot as plt  # Import matplotlib

```

```python
plt.plot(df)  # Plot the data

plt.legend(df.columns, ncol =len(df.columns))  # Add the legend
plt.suptitle(subtitle)  # Add the subtitle
plt.title(title, y =1.125)

plt.show()  # Show the plot
```
